```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/movies.csv")
```

```python
df
```

Средний рейтинг в зависимости от года выпуска

```python
YR=df.groupby('Year')['Rating'].mean()
YR.plot(x = 'Year', y = 'Rating', figsize=( 10 , 8 ), grid = True)
#df.groupby('Year', dropna=False)['Rating'].agg(['count', 'mean'])
```

Преобладание рейтинг MPAA

```python
df.groupby('Certificate')['Year'].agg('count').plot(kind = 'bar',figsize=( 10 , 8 ))
```

Количество фильмов в зависимости от месяца

```python
df.groupby('Month', dropna=False)['Rating'].agg(['count']).plot(kind = 'line', figsize=(10, 8))
```

Рейтинг MMPA в зависимости от месяца

```python
fig, ax = plt.subplots(figsize=(18, 8))
sns.countplot(x = 'Month', hue = 'Certificate', data = df)
plt.legend(loc='upper right')
```

Heatmap


```python
HeatMap=df.pivot_table(index='Month', columns= 'Year', values= 'Rating', aggfunc= ['mean'])
```

```python
sns.heatmap(HeatMap)
```

```python
HeatMap2=df.pivot_table(index='Certificate', columns= 'Month', values= 'Rating', aggfunc= ['count'])
```

```python
HeatMap2
```

```python
sns.heatmap(HeatMap2)
```
